### Ex4: Tính toán trên Array

```r
m1 <- matrix(c(1:9), nrow = 3, ncol = 3, byrow = TRUE)
m2 <- matrix(c(10:18), nrow = 3, ncol = 3, byrow = TRUE)
```

```r
A3 <- array(c(m1,m2), dim = c(3,3,2), dimnames = list(c("r1","r2","r3"), c("cl","c2","c3"),c("matrix1","matrix2")))
print(A3)
```

```r
#In tong theo dong cua tat ca cac ma tran
print("Tong theo dong cua tat ca cac ma tran:")
print(apply(A3, c(1),sum))
print("Tong theo cot cua tat ca cac ma tran:")
print(apply(A3, c(2),sum))
print("Tong theo cac ma tran:")
print(apply(A3, c(3),sum))
print("Tong dong 1 cua matrix 1:")
print(sum(A3[1,,1]))
print("Tich cot 2 cua matrix 2:")
print(prod(A3[,2,2]))
```

### Ex5: Tạo và tính toán trên array

```r
v <- c()
i <-1 
while (i<= 50){
    x = floor(runif(1, min= 5, max = 50))
    if (x%%2 == 0){
        v[i] <- x
        i <- i + 1
    }
}
print(v)
```

```r
#Tao array gom 2 matrix 5x5
A4 <- array(v, dim = c(5,5,2))
print("Array A4:")
print(A4)
```

```r
no_row <- dim(A4)[1]
no_col <- dim(A4)[2]
no_matrix <- dim(A4)[3]
m <- 1
r <- 1
c <- 1
tong <- 0
while(m <= no_matrix){
    r <- 1
    while (r <= no_row){
        c <- 1
        while (c <= no_col){
            if(A4[r,c,m]%%2 == 0 & A4[r,c,m]%%3 == 0 & A4[r,c,m]%%6 == 0){
                print(paste("A4[",r,",",c,",",m,"] =", A4[r,c,m]))
                tong <- tong + A4[r,c,m]
            }
            c <- c +1
        }
       r <- r + 1 
    }
   m <- m + 1 
}
print(paste("Tong =", tong))
```

### Ex7: Tạo và tính toán trên list

```r
#Tao 3 vector
year <- c(2008:2018)
month <- c(1:12)
day <- c(1:31)
```

```r
#Tao list chua 3 vector year, month, day
year_month_day.list <- list("year" = year, "month"=month, "day"=day)
print("List after creating:")
print(year_month_day.list)
```

```r
#Cap nhat year trong list thanh c(2010:2020)
year_month_day.list$year <- c(2010:2020)
print("List after updating:")
print(year_month_day.list)
```

```r
#Bo nam dau tien trong year cua list
year_month_day.list$year <- year_month_day.list$year[-1]
```

```r
print("List after dropping first year:")
print(year_month_day.list)
```

```r
#Tao vector x
x <- c(1,3,4,7,11,18,19)
x.list <- list("x*2"=x*2, "x/2"=x/2, "sqrt(x)"=sqrt(x))
```

```r
print(x.list)
```

```r
#in 3 phan tu giua cua sqrt(x)
mid <- floor(length(x.list$`sqrt(x)`)/2)
print(paste("In 3 phan tu giua cua sqrt(x):", toString(x.list$`sqrt(x)`[mid:(mid+2)])))
```

```r
#Tao list moi chua 2 list x va year_month_day
new.list <- c(year_month_day.list, x.list)
print("new list:")
print(new.list)
```

```r
#tao list con chua year, x*2, day tu tu newlist
sub_new.list <- new.list[c("year","x*2","day")]
print("Sub new list:")
print(sub_new.list)
```

```r
#chieu dai cua element x*2 trong sub_new list
len_element2 <- length(sub_new.list$`x*2`)
print(paste("chieu dai cua element x*2:",len_element2))
```

```r
#In gia tri phan tu cuoi cung trong year
print(paste("Gia tri cua phan tu cuoi cung trong element year:",sub_new.list$year[length(sub_new.list$year)]))
```
