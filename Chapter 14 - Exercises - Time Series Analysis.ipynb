### Ex1: Kings

```r
#Cung cap tap tin kings.dat.txt
#Doc du lieu tu tap tin va in du lieu
kings.dat <- scan("Chapter 14\\kings.dat.txt", skip = 3)
kings.dat
```

```r
#Chuyen du lieu thanh time series object. In Time series object
kings.timeseries <- ts(kings.dat, start = 1, end = length(kings.dat), frequency = 1)
print(kings.timeseries)
```

```r
#Ve timeseries object vua tao
plot.ts(kings.timeseries, main = "Age of Death of Kings", xlab = "Kings", ylab = "Age", col= "blue")
```

### Ex2: Births

```r
#Cung cap tap tin mybirths.dat.txt
births <- scan("Chapter 14\\mybirths.dat.txt", skip = 3)
births
```

```r
#Chuyen du lieu thanh time series object. In Time series object
births.ts <- ts(births, start = c(1946,1), frequency = 12)
print(births.ts)
```

```r
#Ve timeseries object vua tao
plot.ts(births.ts, main = "Births per month in New York city, from Jan 1946 to Dec 1959", xlab = "Year", ylab = "Births", col= "blue")
```

##### Nhận xét: Từ biểu đồ ta thấy biểu đồ có xu hướng năm và từ năm 1950 số trẻ sơ sinh có xu hướng tăng theo năm.

```r
#Thuc hien viec du bao va bieu do so sanh voi thuc tien
comp = decompose(births.ts)
```

```r
comp$x
```

```r
comp$seasonal
```

##### Nhận xét: Tháng 4 có giá trị cao nhất và tháng 11 có giá trị thấp nhất trong yếu tố seasonal -> Tháng 4 có nhiều trẻ sơ sinh nhất trong năm và tháng 11 có ít trẻ sơ sinh nhất trong năm

```r
comp$trend
```

```r
plot(comp)
```

**Nhận xét:** Biểu đồ có xu hướng tăng và chu kỳ theo mùa.

```r
births.ts.forecast <- HoltWinters(births.ts)
print(births.ts.forecast)
```

**Nhận xét:**
- Alpha (0.685 ) khá cao: cho thấy giá trị hiện tại có trọng số khá cao trong việc dự đoán giá trị tương lai.
- Beta (0.0090 khá thấp: cho thấy độ dốc của trend không cập nhật trong thời gian qua.
- gamma (0.68) khá cao: cho thấy ảnh hưởng của yếu tố mùa tại thời điểm hiện tạo phụ thuộc vào giá trị gần nhất.

```r
plot(births.ts.forecast, main = "Births : observation vs forecast", xlab = 'Year', ylab = "Births")
```

```r
#Du doan so luong sinh cho 6 thang tiep theo
births.ts.forecast.6th <- forecast:::forecast.HoltWinters(births.ts.forecast, h = 6)
print("Forecast for next 6 months:")
print(births.ts.forecast.6th)
```

```r
plot(births.ts.forecast.6th, main = "Births : Forecast for next 6 months", xlab = 'Year', ylab = "Births")
```
