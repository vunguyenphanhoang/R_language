### Ex2: Tính tiền điện (điện sinh hoạt)

```r
level1 <- 1678
level2 <- 1734
level3 <- 2014
level4 <- 2536
level5 <- 2834
level6 <- 2927
bac50 <- 50
bac100 <- 100
```

```r
kwh <- as.numeric(readline(prompt = "Moi ban nhap so Kwh dien tieu thu : "))
if (kwh <= 50){
    tongtien <- kwh*level1
} else if (kwh <= 100 ){
    tongtien <-  (bac50*level1) + (kwh - bac50)*level2
} else if (kwh <= 200){
    tongtien <-  (bac50*level1) + (bac50* level2) + (kwh - 2*bac50)*level3
} else if (kwh <= 300){
    tongtien <-  (bac50*level1) + (bac50* level2) + (bac100*level3) + (kwh - 2*bac50 - bac100)*level4
} else if (kwh <= 400){
    tongtien <-  (bac50*level1) + (bac50* level2) + (bac100*level3) + (bac100*level4) + (kwh - 2*bac50 - 2*bac100)*level5
} else {
    tongtien <-  (bac50*level1) + (bac50* level2) + (bac100*level3) + (bac100*level4)+ (bac100*level5) + (kwh - 2*bac50 - 3*bac100)*level6
} 
```

```r
print(paste("So tien dien phai tra : ", format(tongtien, scientific = FALSE, big.mark = ","), "VND"))
```

### Exercise 3: Tính tiền thuê phòng khách sạn

```r
vip1 <- 1000000
vip2 <- 900000
vip3 <- 850000
standard1 <- 700000
standard2 <- 550000
```

```r
room <- as.integer(readline(prompt = " Nhap loai ma loai phong (1/2/3/4/5) = "))
days <- as.integer(readline(prompt = " Nhap so dem luu tru = "))
```

```r
unit_price <- switch(room,vip1, vip2, vip3, standard1, standard2)
```

```r
price <- if(days < 3) {
    price <- unit_price*days
} else if (days <= 6) {
    price <- unit_price *days *0.9
} else if (days >= 7) {
    price <- unit_price *days * 0.8
}
```

```r
print(paste("Tong tien ban can thanh toan = ", format(price, scientific = FALSE, big.mark = ","), "VND"))
```

### Ex4: Trò chơi One-Two-Three

```r
player1 <- readline(prompt = "Moi nguoi choi thu 1 nhap (scissors/rock/paper) : ")
player2 <- readline(prompt = "Moi nguoi choi thu 2 nhap (scissors/rock/paper) : ")
```

```r
if (player1 == player2){
    result <- "The same!"
} else if  (player1 == "scissors"){
    if(player2 == "paper"){
        result <- "Player1 is the winner!"
    } else if ( player2 == "rock"){
        result <- "Player2 is the winner!"
    }
} else if (player1 == "paper"){
    if(player2 == "scissors"){
        result <- "Player2 is the winner!"
    } else if (player2 == "rock"){
        result <- "Player 1 is the winner!"
    }
} else if (player1 == "rock"){
    if(player2 == "scissors"){
        result <- "Player1 is the winner!"
    } else if (player2 == "paper"){
        result <- "Player2 is the winner!"
    }
}
```

```r
print(paste("player 1 : ", player1))
print(paste("player 2 : ", player2))
print(paste(result))
```

### Ex5: Tính cước GOBIKE/GOSEND

```r
dv1_2km <- 10000
dv1_up2km <- 36000
dv2_2km <- 15000
dv2_up2km <- 4000
```

```r
dv <- as.integer(readline(prompt = 'Nhap loai dich vu (1: GOBIKE, 2: GOSEND) : '))
sokm <-  as.numeric(readline(prompt = 'Nhap so km : '))
```

```r
if (dv == 1){
    if (sokm <= 2){
        tongtien <- dv1_2km
    } else {
        tongtien <-  dv1_2km + (sokm - 2)*dv1_up2km
    }
} else {
        if (sokm <=2) {
            tongtien <- dv2_2km
        } else {
            tongtien <- dv2_2km + (sokm -2)*dv2_up2km
        }
    }
```

```r
print(paste('So tien xe can thanh toan : ', format(tongtien, scientific = FALSE, big.mark = ","), 'VND'))
```
